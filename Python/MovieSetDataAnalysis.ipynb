## Exploratory Data Analysis  

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read u.data
ratings_cols = ['user_id','movie_id','rating','timestamp']
df_ratings = pd.read_csv('u.data',sep = '\t', names = ratings_cols)
print(df_ratings)
```

```python
# Read u.item 
movies_cols = ['movie_id','title','release_date','video_release_date','imdb_url']
df_movies = pd.read_csv('u.item',sep = '|',encoding = 'latin-1', 
                       usecols = range(5),names = movies_cols)
pd.reset_option('all')
print(df_movies)
```

```python
# Read u.user
user_columns = ['user_id','age','gender','occupation','zip_code']
df_users = pd.read_csv('u.user',sep='|',encoding='latin-1',names=user_columns)
print(df_users)
```

```python
# Using info and describe, u.data 
print(df_ratings.describe())

print(df_ratings.info())
```

```python
# Using info and describe, u.item
print(df_movies.describe())

print(df_movies.info())
```

```python
# Using info and describe, u.user
print(df_users.describe())

print(df_users.info())
```

```python
# First few rows of the data for each df 
print(df_ratings.head(10))

print(df_movies.head(10))

print(df_users.head(10))
```

```python
# Rows of data for each data frame 

print(df_ratings.shape)
# Rows = 100000

print(df_movies.shape)
# Rows = 1682

print(df_users.shape)
# Rows = 943
```

```python
# unique values and the number of them for rating
print(df_ratings['rating'].unique())
print(df_ratings['rating'].nunique())
```

```python
# unique values and the number of them for user_id
print(df_ratings['user_id'].unique())
print(df_ratings['user_id'].nunique())

# The number of unique values 943
# The number means that there are 943 users who 
# gave their rating based on a movie based on the movie id
```

```python
# Historgram for the number of times each rating was used 
rating_counts = df_ratings['rating'].values

plt.hist(rating_counts, bins = 50)

plt.show()
```

```python
# Movie rated the greatest number of times
movie_ratings_count = df_ratings.groupby('movie_id')['rating'].count()

most_rated_movie_id = movie_ratings_count.idxmax()

most_rated_movie_title = df_movies[df_movies['movie_id'] == most_rated_movie_id]['title'].values[0]

print(f"The movie that was rated the greatest number of times is: {most_rated_movie_title}")
```

```python
# Id of the movie with the most 5 star rating
five_star_ratings = df_ratings[df_ratings['rating'] == 5]

five_star_movie_ratings_count = five_star_ratings.groupby('movie_id')['rating'].count()

most_rated_movie_id = five_star_movie_ratings_count.idxmax()

print(f"The ID of the movie that received the most '5' ratings is: {most_rated_movie_id}")
```

```python
# Histogram of the frequency of ratings by user
user_ratings_count = df_ratings.groupby('user_id')['rating'].count()

plt.figure(figsize=(10,6))
plt.hist(user_ratings_count, bins=50,)
plt.title('Frequency of Ratings by User')
plt.xlabel('Number of Ratings')
plt.ylabel('Number of Users')
plt.show()
```

```python
# Date with the greatest number of movies released
df_movies['release_date'] = pd.to_datetime(df_movies['release_date'])

movies_release_count = df_movies.groupby('release_date').size()

most_movies_released_date = movies_release_count.idxmax()

print(f"The date with the greatest number of movies released is: {most_movies_released_date.strftime('%Y-%m-%d')}")
```

```python
# Merging dataframes for ratings and movie data
df_merged = pd.merge(df_ratings,df_movies,on='movie_id')

print(df_merged.sample(5))
```

```python
# Star Wars rated 
star_wars_ratings = df_merged[df_merged['title'] == 'Star Wars (1977)']

num_ratings = len(star_wars_ratings)

print(f"The movie 'Star Wars (1977)' was rated {num_ratings} times.")
```

```python
# Histogram using the number of ratings per movie
ratings_per_movie = df_merged.groupby('title')['rating'].count()

plt.figure(figsize=(10,6))
plt.hist(ratings_per_movie, bins=50)
plt.title('Number of Ratings per Movie')
plt.xlabel('Number of Ratings')
plt.ylabel('Number of Movies')
plt.show()
```

```python
# Bar char of the top-rated movies
average_ratings = df_merged.groupby('title')['rating'].mean()

top_rated_movies = average_ratings.sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
top_rated_movies.plot(kind='bar')
plt.title('Top-Rated Movies')
plt.xlabel('Movie Title')
plt.ylabel('Average Rating')
plt.show()
```

```python
# merge user df with pervious merge df 
df_final = pd.merge(df_merged, df_users, on='user_id')

print(df_final.sample(5))
```

```python
# Number of ratings and average rating for each movie
movie_stats = df_final.groupby('title').agg({'rating': ['count', 'mean']})

print(movie_stats)
```

```python
# Top-rated movies with a min rating count of 100
popular_movies = movie_stats[movie_stats[('rating', 'count')] >= 100]

top_rated_popular_movies = popular_movies.sort_values(('rating','mean'), ascending=False)

print(top_rated_popular_movies)
```

```python
# Average rating for each movie by gender and a histogram for each gender
average_ratings_by_gender = df_final.groupby(['title', 'gender'])['rating'].mean().unstack()

plt.figure(figsize=(10,6))
plt.hist([average_ratings_by_gender['F'].dropna(), average_ratings_by_gender['M'].dropna()], 
         bins=50,label=['Female', 'Male'])
plt.title('Average Movie Ratings by Gender')
plt.xlabel('Average Rating')
plt.ylabel('Number of Movies')
plt.legend(loc='upper left')
plt.show()
```

```python
# List the number of people in each job in descending order
occupation_counts = df_users['occupation'].value_counts()

sorted_occupation_counts = occupation_counts.sort_values(ascending=False)

print(sorted_occupation_counts)
```

```python
# Bar char of occupations
occupation_counts = df_users['occupation'].value_counts()

plt.figure(figsize=(10,6))
occupation_counts.plot(kind='bar')
plt.title('Number of People in Each Occupation')
plt.xlabel('Occupation')
plt.ylabel('Number of People')
plt.show()
```

```python
# Boxplot for age
plt.figure(figsize=(10,6))
df_users['age'].plot(kind='box')
plt.title('Boxplot of Age')
plt.ylabel('Age')
plt.show()
```
